```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import io


np.random.seed(42)
```

```python
# model : [body_type, base_price]
cars = {
    'Toyota Agya': ['Hatchback', np.random.randint(100,200) * 1e6],
    'Toyota Yaris': ['Hatchback', np.random.randint(100,200) * 1e6],
    'Toyota Calya': ['Wagon', np.random.randint(100,200) * 1e6],
    'Daihatsu Alya': ['Hatchback', np.random.randint(100,200) * 1e6],
    'Daihatsu Terios': ['SUV', np.random.randint(100,200) * 1e6],
    'Daihatsu Xenia': ['MPV', np.random.randint(100,200) * 1e6],
    'Honda Jazz': ['Hatchback', np.random.randint(100,200) * 1e6],
    'Honda CR-V':['SUV', np.random.randint(100,200) * 1e6],
    'Honda Civic': ['Sedan', np.random.randint(100,200) * 1e6],
    'Suzuki Ertiga': ['MPV', np.random.randint(100,200) * 1e6]
}
cars
```

```python
products_dummy = {}
count = 0
for car, (body_type, _) in cars.items():
  how_many_types_available = np.random.randint(3,7)
  years = np.random.choice([y + 2014 for y in range(7)], size = how_many_types_available, replace = False)
  for year in np.sort(years):
    count += 1
    products_dummy.setdefault('product_id', []).append(count)
    products_dummy.setdefault('merk', []).append(car.split()[0])
    products_dummy.setdefault('model', []).append(car)
    products_dummy.setdefault('body_type', []).append(body_type)
    products_dummy.setdefault('transmission_type', []).append(np.random.choice(['Automatic', 'Manual']))
    products_dummy.setdefault('year', []).append(year)
```

```python
products = pd.DataFrame(products_dummy)
products.to_csv('/content/products.csv', index=False)
products
```

```python
def generate_mock_phone_number():
  """Generates a random 10-digit phone number in a specific format."""
  area_code = str(np.random.randint(100, 999))
  prefix = str(np.random.randint(100, 999))
  line_number = str(np.random.randint(1000, 9999))
  return f"{area_code}-{prefix}-{line_number}"
```

```python
import names

all_names = [names.get_full_name() for _ in range(100)]
```

```python
sellers = {}
seller_name = all_names[:25]
for i in range(25):
  sellers.setdefault('seller_id', []).append(i+1)
  sellers.setdefault('name', []).append(seller_name[i])
  sellers.setdefault('phone', []).append(generate_mock_phone_number())
  sellers.setdefault('city_id', []).append(np.random.choice(cities['city_id']))
sellers = pd.DataFrame(sellers)
sellers.to_csv('/content/sellers.csv', index=False)
```

```python
buyers = {}
buyer_name = all_names[25:]
for i in range(75):
  buyers.setdefault('buyer_id', []).append(i+1)
  buyers.setdefault('name', []).append(buyer_name[i])
  buyers.setdefault('phone', []).append(generate_mock_phone_number())
  buyers.setdefault('city_id', []).append(np.random.choice(cities['city_id']))
buyers = pd.DataFrame(buyers)
buyers.to_csv('/content/buyers.csv', index=False)
```

```python
product_onsales = {}
count = 0
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 6, 30)
for i, seller in enumerate(seller_name):
  how_many_cars_sold_by_seller = np.random.randint(1, 4)
  for numb in range(how_many_cars_sold_by_seller):
    count += 1
    product_onsales.setdefault('sales_id', []).append(count)
    product_onsales.setdefault('seller_id', []).append(i+1)

    product_id_sold_by_seller = np.random.choice(products['product_id'])
    product_onsales.setdefault('product_id', []).append(product_id_sold_by_seller)

    car_year = products.loc[product_id_sold_by_seller-1, 'year']
    car_model = products.loc[product_id_sold_by_seller-1, 'model']
    base_price = cars[car_model][1]
    car_price = base_price + round((np.random.rand()/5) + (car_year-2014), 3) * 1e7
    product_onsales.setdefault('price', []).append(int(car_price))
    product_onsales.setdefault('is_negotiable', []).append(True)

    random_date = start_date + timedelta(days=np.random.randint(0, (end_date - start_date).days))
    product_onsales.setdefault('date_post', []).append(random_date)

product_onsales = pd.DataFrame(product_onsales)
product_onsales.to_csv('/content/product_onsales.csv', index=False)
product_onsales
```

```python
bids = {}
count = 0

for i, buyer in enumerate(buyer_name):
  how_many_car_bidden = np.random.randint(1,4)
  sale_will_be_bidden = np.random.choice(product_onsales['sales_id'], size=how_many_car_bidden)

  for sales in sale_will_be_bidden:
    how_many_bids_per_sales = np.random.randint(1,3)
    date_bid = np.empty(how_many_bids_per_sales, dtype=object)
    bid_prices = np.empty(how_many_bids_per_sales, dtype=int)

    date_post = product_onsales.loc[sales-1,'date_post'].date()
    sales_price = product_onsales.loc[sales-1,'price']

    for j in range(how_many_bids_per_sales):
      count += 1
      bids.setdefault('bid_id', []).append(count)
      bids.setdefault('sales_id', []).append(sales)
      bids.setdefault('buyer_id', []).append(i+1)

      if j == 0:
        date_bid[j] = date_post + timedelta(days=np.random.randint(30))
        bid_prices[j] = sales_price - round(np.random.randint(5, 20)) * 1e6
      else:
        date_bid[j] = date_bid[j-1] + timedelta(days=np.random.randint(10))
        bid_prices[j] = bid_prices[j-1] - round(np.random.randint(5, 20)) * 1e6

      bids.setdefault('date_bid', []).append(date_bid[j])
      bids.setdefault('bid_price', []).append(bid_prices[j])

bids = pd.DataFrame(bids)
bids.to_csv('/content/bids.csv', index=False)
bids
```
